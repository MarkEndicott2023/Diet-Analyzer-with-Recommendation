```python
import sympy as sp
import numpy as np

# Store apple nutrients in a matrix.
# Columns: Protein, Fats, Carbs

apple_data = ['Apple', 100, sp.Matrix(np.array([.26,.17,13.8]))]
orange_data = ['Orange', 100, sp.Matrix(np.array([.91,.15,10.6]))]
banana_data = ['Banana', 100, sp.Matrix(np.array([.74, .29, 23]))]

food_data = [apple_data, orange_data, banana_data]
```

```python

'''Part 1 functions here'''

############################################################################################################

def get_weight_info():

    # Prompts the user for weight, ideal weight, and desired weight loss per week.
  
    weight = float(input("Enter your current weight (lbs): "))
    ideal_weight = float(input("Enter your ideal weight (lbs): "))
    weight_loss = float(input("Enter desired pounds lost per week: "))
    date_loss = float(input("Enter number of weeks until desired diet end: "))
    
    return weight, ideal_weight, weight_loss, date_loss

############################################################################################################

def display_food_list(food_data):
    
    # Displays a list of foods in the food data dictionary.
  
    print('''\nAvailable Foods:\n\nThe following data was extracted from the U.S. Depatment of Agriculture \n''')
    
    for i, food in enumerate(food_data):
        print(f"{i+1}.\n \nFood: {food[0]} \nGrams: {food[1]}")
        print("\nRespective Macros: ")
        print("\n Proteins \n Fats \n Carbs \n")
        sp.pprint(food[2])
        print()

############################################################################################################

def get_food_intake(food_data):
    
    # Prompts the user for food type and quantity and scales the food matrix.
    # Stores user information in a variable.

    end_key = 0  
    user_mat = []

    while end_key != "Yes":

        # Prompts for user's food information.

        print("Greetings! Enter your food today as specified below.\n")
        food_type = str(input("Enter your food name as shown: "))
        food_quantity = int(input("Enter the quantity of your food item: "))
        end_key = str(input("Are you done? (Type Yes or No) "))
        print()

        # Scaling and storing user intake.

        for item in food_data:

            if food_type == item[0]:

                food_mat = item[2] * int(food_quantity)
                user_mat.append([item[0], food_mat])

    # Display results

    print("\nRespective Macros: ")
    print("\n Proteins \n Fats \n Carbs \n")

    sp.pprint(user_mat)
    
    return user_mat

############################################################################################################

def calculate_cal(total_user_intake):
    
    # Uses matrix operations to calculate total macro and calorie count.

    cal_gram = sp.Matrix(np.array([[4,4,9]]))
    cal_list = []

    for item in total_user_intake:

        cal_list.append(cal_gram @ item[1])

    result = sp.Matrix.zeros(1, 1)

    for matrix in cal_list:
        result += matrix
    
    print("\nCalorie Estimate: \n")
    sp.pprint(result)
    
    return result
```

```python
'''Part 2 Functions Here'''

############################################################################################################

def display_cal_deficits (weight, ideal_weight, weight_loss, date_loss):

    # Display calorie calculations based on weight data.
    
    calories_pound = 3500

    weight_diff = weight - ideal_weight
    total_cal_to_lose = weight_diff * calories_pound
    weekly_deficit = total_cal_to_lose / date_loss
    daily_deficit = weekly_deficit / 7

    print("From your input, the following calorie information was calculated: \n",
          "\nWeight: ", weight, 
          "\nWeight Goal: ", ideal_weight, 
          "\nDesired Pound Loss Per Week: ", weight_loss, 
          "\nPounds to lose: ", weight_diff,
          "\n",
          "\nCalories to burn: ", total_cal_to_lose,
          "\nCalorie deficit per week: ", weekly_deficit,
          "\nCalorie deficit per day: ", daily_deficit)
    
    return daily_deficit
    
############################################################################################################

def macro_percentages(total_user_intake):

    carb_perc_range = range(45,65)
    prot_perc_range = range(10,35)
    fat_perc_range = range(20,35)

    total_lst = list(total_user_intake)
    T = 0

    for item in total_lst:

        print("\n\nMacro Percentage in Protein, Fat, and Carbs: ")

        for macro in item[1]:

            T += 1
            
            food_thermos = [4,4,9]
            perc_macro = round((macro * food_thermos[T-1]) / int(cal_result[0]) * 100)
            print('\n', str(perc_macro) + '%')

            if T == 1:
                prot_perc = perc_macro
                if (perc_macro in prot_perc_range): print("balanced")
                else: print("imbalanced Protein")    

            if T == 2:
                fat_perc = perc_macro
                if (perc_macro in fat_perc_range): print("balanced")
                else: print("imbalanced Fat")  

            if T == 3:
                carb_perc = perc_macro
                if (perc_macro in carb_perc_range): print("balanced")
                else: print("imbalanced Carbs")

        return prot_perc, fat_perc, carb_perc

############################################################################################################

def macro_comparison(prot_perc, fat_perc, carb_perc):

    # Prints a comparison between the recommended macro ranges and user data for the day.

    carb_perc_range = range(45,65)
    prot_perc_range = range(10,35)
    fat_perc_range = range(20,35)
    prot_diff = 0
    fat_diff = 0
    carb_diff = 0
    
    print("\nThe following ranges are the macronutrient recommended daily intake from the USDA:\n",
         "\nProtein: 10-35%",
         "\nFat: 20-35%",
         "\nCarbohydrates: 45-65%",
         "\n\nThe following numbers represent how far you are from hitting the minimum or maximum:\n")

    # Protein

    if prot_perc < min(prot_perc_range): prot_diff = prot_perc - min(prot_perc_range)
    elif prot_perc > max(prot_perc_range): prot_diff = prot_perc - max(prot_perc_range)

    if prot_diff < 0: print(str(prot_diff) + '%', ", Need more proteins")
    elif prot_diff > 0: print(str(prot_diff) + '%', ", Need less proteins")
    elif fat_diff == 0: print("Proteins Within Range")

    # Fat

    if fat_perc < min(fat_perc_range): fat_diff = fat_perc - min(fat_perc_range)
    elif fat_perc > max(fat_perc_range): fat_diff = fat_perc - max(fat_perc_range)

    if fat_diff < 0: print(str(fat_diff) + '%', ", Need more fats")
    elif fat_diff > 0: print(str(fat_diff) + '%', ", Need less fats")
    elif fat_diff == 0: print("Fats Within Range")
    
    # Carbs

    if carb_perc < min(carb_perc_range): carb_diff = carb_perc - min(carb_perc_range)
    elif carb_perc > max(carb_perc_range): carb_diff = carb_perc - max(carb_perc_range)
    
        
    if carb_diff < 0: print(str(carb_diff) + '%', ", Need more carbs")
    elif carb_diff > 0: print(str(carb_diff) + '%', ", Need less carbs")
    elif carb_diff == 0: print("Carbs Within Range")
        
    return prot_diff, fat_diff, carb_diff
```
